```python
#que 1
import seaborn as sns
import plotly.express as px

# Load the "titanic" dataset
titanic_dataset = sns.load_dataset("titanic")

# Plot scatter plot using Plotly Express
scatter_plot = px.scatter(titanic_dataset, x='age', y='fare', title='Scatter Plot of Age vs Fare')

# Show the scatter plot
scatter_plot.show()
```

```python
#que 2
import plotly.express as px

# Load the "tips" dataset from Plotly
tips_dataset = px.data.tips()

# Plot the box plot using Plotly Express
box_plot = px.box(tips_dataset, x="day", y="total_bill", title="Box Plot of Total Bill by Day")

# Show the box plot
box_plot.show()
```

```python
#que 3
import plotly.express as px

# Load the "tips" dataset from Plotly
tips_dataset = px.data.tips()

# Plot the histogram using Plotly Express
histogram = px.histogram(tips_dataset, x="sex", y="total_bill", color="day", pattern_shape="smoker",
                         title="Histogram of Total Bill by Sex", barmode="group")

# Show the histogram
histogram.show()
```

```python
#que 4
import pandas as pd
import plotly.express as px

# Load the "iris" dataset from Plotly
iris_dataset = px.data.iris()

# Create a scatter matrix plot using pandas plotting function
scatter_matrix = pd.plotting.scatter_matrix(iris_dataset, color=iris_dataset['species'].map({'setosa': 'red', 'versicolor': 'green', 'virginica': 'blue'}))

# Convert the scatter matrix plot to a Plotly figure
scatter_matrix_fig = px.scatter_matrix(iris_dataset, color=iris_dataset['species'], title="Scatter Matrix Plot of Iris Dataset")

# Show the scatter matrix plot
scatter_matrix_fig.show()
```

```python
#que 5
import plotly.express as px

# Load the "iris" dataset from Plotly
iris_dataset = px.data.iris()

# Plot a distribution plot (distplot)
distplot = px.histogram(iris_dataset, x="sepal_width", marginal="rug", title="Distribution Plot of Sepal Width")

# Show the distribution plot
distplot.show()
```
